```python
import duckdb
import pandas as pd
```

## Creating Tables and Running all models in Development

```python
sql_query = """
show tables
"""
# connect to the DuckDB database
with duckdb.connect('./data/nyc_parking_violations.db') as con:
    display(con.execute(sql_query).df())
```

```python
sql_query = """
SELECT * FROM silver_parking_violation_tickets
"""
# connect to the DuckDB database
with duckdb.connect('./data/nyc_parking_violations.db') as con:
    display(con.execute(sql_query).df())
```

```python
# creating table and loading data
sql_query1 = """
CREATE or REPLACE TABLE nyc_parking_violation_codes AS
SELECT * FROM read_csv_auto('./data/ParkingViolationCodes_January2020.csv', normalize_names = True);
"""

sql_query2 = """
CREATE or REPLACE TABLE nyc_parking_violations_2023 AS
SELECT * FROM read_csv_auto('./data/Parking_Violations.csv', normalize_names =True, ignore_errors = True);
"""

with duckdb.connect('./data/nyc_parking_violations.db') as conn:
    conn.sql(sql_query1)
    conn.sql(sql_query2)
```
